公司目前汇总了6月以来12个部门每周的周报，按照提交日期分开，共48个Excel表格。

每张表格包含了八个字段：

- 部门
- 日期
- 工作内容
- 是否完成
- 延期天数
- 责任人
- 本周进度
- 状态

我们需要分别按照部门汇总合并周报表格，以做留存整理。统计每人月度完成情况，部门整体完成率，以及公司本月目标完成情况。

## 查看文件

在对周报进行操作前，我们想先查看一下共有哪些表格。定义一个函数，实现的功能是扫描指定路径下的表格文件。

```python
import os
'''
获取文件夹的路径,该路径下的所有文件夹，以及所有文件
root    # 当前目录路径  
dirs    # 当前路径下所有子目录  
files   # 当前路径下所有非目录子文件 
'''

def get_allfile_msg(file_dir):
    for _, _, files in os.walk(file_dir):
        return [file for file in files if file.endswith('.xls') or file.endswith('.xlsx')]
files = get_allfile_msg(r'weekly_report/input')
```

```python
files = get_allfile_msg(r'weekly_report/input')
files
```

```python
import pandas as pd

df = pd.read_excel('weekly_report/input/2020-6-7财务部周报.xlsx', index = False)
df.head()
```

下面我们把48个Excel表格合并在一张大表中，方便统计每个人的任务完成情况。

```python
df_all = pd.DataFrame()
for file in files:
    df_all = pd.concat([df_all, pd.read_excel('./weekly_report/input/' + file)])
    
df_all
```

下面来查看一下这张大表中都包含哪些日期和部门。

```python
df_all['日期'].value_counts().index
```

```python
df_all['部门'].value_counts().index
```

## 个人任务情况

第一步先根据周报内容统计每个员工的任务完成情况。根据领导的指示，需要统计出6月的四周中，每周都按时完成任务的员工名单，以及超过三周未完成任务名单。

### 按部门+责任人排序

刚才合并好的表格中，直接把各部门不同日期的表格拼接在一起，没有做数据顺序上的处理。现在要统计个人任务完成状况，需要按照“部门”和“责任人”排序，将同一员工四周的周报内容放在相邻的几行中，这样就能非常清晰地查看每个人的周报内容和任务情况。

```python
df_sort = df_all.sort_values(by=['部门', '责任人']).reset_index(drop=True)
df_sort
```

```python
df_sort.to_excel('./weekly_report/个人任务完成情况.xlsx', index=False)
```

### 未完成次数大于等于3的员工统计

根据“是否完成”一列的数据，在原表格中新增一列，命名为“是否完成bool”。

```python
df_sort['是否完成bool'] = df_sort['是否完成'].apply(lambda x: True if x == '是' else False)
df_sort
```

按照部门和责任人分组，统计每个员工“是否完成bool”一列值为“False”的个数（4-True的数量），就可以得到6月中未完成任务次数。

```python
df_delay = 4 - df_sort.groupby(['部门', '责任人'])['是否完成bool'].sum()
df_delay
```

这样就统计到了所有员工的未完成任务的次数。最后筛选出延期次数大于等于3的员工名单。

```python
df_delay[df_delay >= 3]
```

### 四次均按时完成的员工统计

```python
df_ontime = df_delay[df_delay == 0]
df_ontime
```

```python
df_ontime.to_excel('./weekly_report/按时完成名单.xlsx')
```

## 部门任务完成情况

### 每周完成率

首先按照“部门”和“日期”分组，统计每周各部门任务完成率。

```python
department_group_week = df_sort.groupby(['部门', '日期'])['是否完成bool']
df_department_week_complete_rate = department_group_week.sum() / department_group_week.count()
df_department_week_complete_rate
```

### 整体完成率

再来计算部门月度整体完成率。

```python
department_group_month = df_sort.groupby(['部门'])['是否完成bool']
df_department_month_complete_rate = department_group_month.sum() / department_group_month.count()
df_department_month_complete_rate
```

人事部、兼职部、研发部、财务部、转化部都100%完成了6月计划，运营部、销售部、教学部还需要提高完成率，产品部、管理部的完成率最低。

## 公司任务完成情况

### 每周完成率

```python
company_group_week = df_sort.groupby(['日期'])['是否完成bool']
df_company_week_complete_rate = company_group_week.sum() / company_group_week.count()
df_company_week_complete_rate
```

可以看到公司6月每周任务完成率大概稳定在80%左右。

### 整体完成率

```python
df_company_month_complete_rate = df_sort['是否完成bool'].sum() / df_sort['是否完成bool'].count()
df_company_month_complete_rate
```

## 按部门生成工作表

为了让每月周报能更有序地留存，我们把所有周报按照部门生成到同一Excel的不同工作表中。只需要按照“部门”划分数据，再放置到相应的工作表中就可以了，同时删除掉我们刚才新增的“是否完成bool”一列的数据。

```python
departments = df_sort['部门'].value_counts().index

writer = pd.ExcelWriter('./weekly_report/按部门分页.xlsx')
for department in departments:
    df_dep = df_sort[df_sort['部门'] == department].copy()
    df_dep.drop('是否完成bool', inplace=True, axis=1)
    df_dep.to_excel(writer, department, index=False)

writer.save()
```
